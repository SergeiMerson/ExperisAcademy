# Two Matplotlib Interfaces

### MATLAB-Style Interface

```python
plt.subplot(2,1,1)
plt.plot(x, np.sin(x))

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))

plt.show()
```

### Object-Oriented Interface

```python
fig, ax = plt.subplots(2)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

plt.show()
```

```python
S = np.linspace(0, 1, 1000)
Km = 0.04
Vmax = 0.1

v = (Vmax * S) / (Km + S)

fig, ax = plt.subplots(figsize=(8,6))

fig.patch.set_facecolor('blue')
fig.patch.set_alpha(0.2)

ax.plot(S, v)
ax.set(
       title='The Michaelis-Menten equation',
       xlabel='S', ylabel='v',
       facecolor='yellow')
ax.patch.set_alpha(0.5)
ax.grid(axis='y')
ax.axis('tight')
plt.show()
```

### Moore’s Law Exercise

```python
obs_x = np.array([72,74,78,82,85,89,93,97,99,100,103,104,107,108,112], dtype='uint16') + 1900
obs_x
```

```python
obs_y = np.array([0.0025, 0.005, 0.029, 0.12, 0.275, 1.18, 3.1, 7.5, 24, 42, 220, 592, 1720, 2046, 3100])
obs_y
```

```python
obs_y.shape == obs_x.shape
```

```python
pred_x = np.arange(1972, 2015, 5, dtype='uint16')
pred_x
```

```python
x_0 = obs_x[0]
y_0 = obs_y[0]
pred_y = y_0 * 2 ** ((pred_x - x_0) / 2)
```

```python
fig, ax = plt.subplots(figsize=(9,6))

ax.plot(pred_x, pred_y, '--k', label='predicted')
ax.plot(obs_x, obs_y, '*r', markersize=10, label='observed')

ax.set(yscale='log',
       title='Moore\'s Law',
       xlabel='Years',
       ylabel='log(n trans)',
       xticks=pred_x,
       xticklabels=pred_x
      )
ax.grid()
# ax.tick_params(axis='x', rotation=45)
ax.autoscale(enable=True, axis='both', tight=True)
ax.axis('tight')
ax.legend(loc='upper left')
plt.show()
```

```python
n0 = 0.0025 * 10**6
y0 = 1972

years = np.arange(y0, 2013)
observed_years = np.array([1972, 1974, 1978, 1982, 1985, 1989, 1993, 1997, 1999, 2000, 2003, 2004, 2007, 2008, 2012])
observed_trans = np.array([0.0025, 0.005, 0.029, 0.12, 0.275, 1.18, 3.1, 7.5, 24, 42, 220, 592, 1720, 2046, 3100])

observed_trans = np.log10(observed_trans * 10 ** 6)
predictions = np.log10(n0) + (years - y0)/2 * np.log10(2)

fig, ax = plt.subplots(figsize=(12,7))

ax.plot(years.tolist(), predictions, '--k', label='prediction')
ax.plot(observed_years.tolist(), observed_trans, '*r', label='observed', markersize=10)

ax.grid()
ax.legend(numpoints=1, loc=0)

ax.set(title='Moore\'s Law', xlabel='Years', ylabel='log(ntrans)')

plt.show()
```
