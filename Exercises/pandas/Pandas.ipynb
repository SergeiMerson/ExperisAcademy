```python
population_df = pd.melt(
    pop_df.drop(
        columns=['Indicator Name', 'Indicator Code']
    ),
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name='Population')


population_df.head()
```

```python
df = population_df
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.loc[:, 'Population'] = (df.Population / 1000).astype('uint32')
```

```python
israel = df.loc[ df['Country Name'] == 'Israel', : ].copy()
```

```python
israel.head()
```

```python
israel.set_index('Year', inplace=True)
israel.drop(columns=['Country Name', 'Country Code'], inplace=True)
```

```python
israel.head()
```

```python
israel.loc[:, 'Change %'] = israel.Population.pct_change()
```

```python
israel.head()
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(2,1, figsize=(16,8), sharex=True)

pop = ax[0].plot(israel.index, israel['Population'])
pct = ax[1].plot(israel.index, israel['Change %']*1000)
plt.xticks(rotation=90)
plt.show()
```

##### parse ingredients

#### Recipes with exactly 3 ingredients

```python
rec_df[rec_df.ingr_N == 3].shape[0]
```

```python
def has_ingredients(all_ing_list, ing_on_hand):
    return all(
                (any(
                        (ing_on.lower() in elem for elem in all_ing_list)

                
                ) for ing_on in ing_on_hand)
              )
```

```python
test_list_result = rec_df[rec_df.ingr.apply(lambda ing_list: has_ingredients(ing_list, ['parsley','paprika','tarragon']))]
test_list_result
```
