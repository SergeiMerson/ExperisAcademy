# Random

```python
import numpy as np
```

```python
np.random.seed(0)
np.random.randint(low=10, high=100, size=(6,15))
```

### random floating point numbers

```python
np.random.random_sample((5,5))
```

```python
np.random.rand(5,5)
```

```python
a, b = 10, 20
data = a + (b - a) * np.random.random_sample(100000)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1,1, figsize=(16,6))
ax.hist(data, bins=100, color='darkred', rwidth=0.5)
plt.xlim(10,20)
plt.show()
```

```python
average_waiting_time = [(10 + 10*np.random.random_sample(100)).mean() for _ in range(100000)]
```

```python
fig = plt.figure(figsize=(16,6))
plt.hist(average_waiting_time, bins=100, color='navy', rwidth=0.5)
# plt.xlim(10,20)
plt.show()
```

### choice

```python
np.random.choice(a=average_waiting_time, size=5, replace=True)
```

```python
np.random.choice(a=[1,2,3,4,5], size=20, replace=True, p=[0.5,0.3,0.1,0.05,0.05])
```

# Random Permutations

There are two methods for permuting the contents of an array:
* np.random.shuffle() randomly rearranges the order of the elements in place
* np.random.permutation() makes a copy of the array first, leaving the original unchanged

```python
arr = np.random.randint(1,1000, (2,5,10))
arr
```

```python
for row in range(arr.shape[1]):
    np.random.shuffle(arr[0,row,:])
```

# The Normal Distribution

```python
np.random.normal(loc=10.0, scale=10, size=(2,3,4))
```

##### It is also possible to draw numbers from the standard normal distribution (that with  = 0 and  = 1) with np.random.randn()
* Like random.rand, this takes the dimensions of an array as its arguments

```python
np.random.randn(2,3,4)
```

```python
waiting_time = np.random.normal(loc=10, scale=3, size=1000000)
```

```python
fig = plt.figure(figsize=(16,6))
plt.hist(waiting_time, bins=10000, color='slategray', rwidth=0.5, histtype='stepfilled')
# plt.xlim(10,20)
plt.show()
```

# The Binomial Distribution

# The Poisson Distribution

```python
np.random.poisson(lam=10, size=(5,5))
```

```python
arr_poisson = np.random.poisson(lam=10, size=1000000)
```

```python
fig = plt.figure(figsize=(16,6))
plt.hist(arr_poisson, bins=20, color='brown', rwidth=0.9)
plt.xlim(0)
plt.show()
```
