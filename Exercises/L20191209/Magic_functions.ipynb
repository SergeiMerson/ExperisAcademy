# %timeit and %%timeit

```python
import random


numbers = list(range(1,100))
random.shuffle(numbers)
```

```python
%timeit sorted(numbers)
```

```python
%timeit -r 1 sorted(numbers)
```

```python
%timeit -r 1000 -n 1000 sorted(numbers)
```

```python
%timeit -r 20 sorted(numbers)
```

```python
%%timeit

n=150
factors = set()

for i in range(1, n+1):
    if n % i == 0:
        factors.add(n)
```

### Exercise

```python
import math
```

```python
%%timeit

n=150
factors = set()

for i in range(1, int(math.sqrt(n))+1):
    if n % i == 0:
        factors.add(n)
```

```python
def generator(n):
    for i in range(1, int(math.sqrt(n))):
        if n % i == 0:
            yield i
```

```python
%%timeit

n=150
factors = set()

for i in generator(n):
    factors.add(i)
```
