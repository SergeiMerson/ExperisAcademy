```python
import numpy as np
```

```python
def get_saddle_points(arr):
    l = arr.shape[0]
    min0_max1 = (np.min(arr, axis=0) == arr) & (np.max(arr, axis=1).reshape(l,1) == arr)
    min1_max0 = (np.max(arr, axis=0) == arr) & (np.min(arr, axis=1).reshape(l,1) == arr)
    return arr[min0_max1 | min1_max0]
```

```python
arr = np.array([[10, 12,  7,  3, 12],
                [3 , 10,  6,  2,  8],                
                [18, 24 ,17,  6, 10],
                [15, 21, 10,  8, 12],
                [1,  18, 22,  4, 15]])
```

```python
get_saddle_points(arr)
```

```python
arr2 = np.array([[10, 12,  7,  3, 12],
                 [3 , 10,  6,  2,  8],
                 [15, 21, 10,  8, 12],
                 [1,  18, 22,  4, 15],
                 [18, 24 ,17,  6, 10]])
```

```python
get_saddle_points(arr2)
```
